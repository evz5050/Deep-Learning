## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
```

```python


#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("charity_data.csv")
application_df.head()
```

```python
application_df.shape
```

```python
# Drop the non-beneficial columns
#  YOUR CODE GOES HERE

application_df = application_df.drop(columns=['EIN', 'STATUS', 'SPECIAL_CONSIDERATIONS'])
application_df
```

```python
# Determine the number of unique values in each column.
#  YOUR CODE GOES HERE

print(application_df.nunique())
```

```python
# name
name_count = application_df['NAME'].value_counts()
name_count
```

```python
name_replace = list(name_count[name_count<100].index)

# Replace in dataframe
for name in name_replace:
    application_df['NAME'] = application_df['NAME'].replace(name,"Other")

# Check to make sure binning was successful
application_df['NAME'].value_counts()
```

```python
# Look at APPLICATION_TYPE value counts for binning
#  YOUR CODE GOES HERE

application_type_value_counts = application_df['APPLICATION_TYPE'].value_counts()
application_type_value_counts
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
#  YOUR CODE GOES HERE

application_types_to_replace = list(application_type_value_counts[application_type_value_counts<100].index)

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
#  YOUR CODE GOES HERE

classification_value_counts = application_df['CLASSIFICATION'].value_counts()
classification_value_counts
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
#  YOUR CODE GOES HERE

xxx = application_df['CLASSIFICATION'].value_counts()[application_df['CLASSIFICATION'].value_counts() > 1]
xxx
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
#  YOUR CODE GOES HERE

classifications_to_replace = list(classification_value_counts[classification_value_counts<100].index)

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# organization

organization_count = application_df['ORGANIZATION'].value_counts()

organization_replace = list(organization_count[organization_count<500].index)
for org in organization_replace:
    application_df['ORGANIZATION'] = application_df['ORGANIZATION'].replace(org,"Other")
application_df['ORGANIZATION'].value_counts()
```

```python
# Use case

use_case_count = application_df['USE_CASE'].value_counts()
use_case_replace = list(use_case_count[use_case_count<500].index)
for use_case in use_case_replace:
    application_df['USE_CASE'] = application_df['USE_CASE'].replace(use_case,"Other")
application_df['USE_CASE'].value_counts()
```

```python
# affliation

affliation_count = application_df['AFFILIATION'].value_counts()

affliation_replace = list(affliation_count[affliation_count<500].index)
for affliation in affliation_replace:
    application_df['AFFILIATION'] = application_df['AFFILIATION'].replace(affliation,"Other")
application_df['AFFILIATION'].value_counts()
```

```python
# income count
income1 = ["1M-5M", "5M-10M", "10M-50M", "50M+"]
for income in income1:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(income, "1M+")
income2=["1-9999", "10000-24999"]
for income in income2:
    application_df['INCOME_AMT'] = application_df['INCOME_AMT'].replace(income, "1-24999")


income_count =  application_df['INCOME_AMT'].value_counts()
income_count
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
#  YOUR CODE GOES HERE

dummies = pd.get_dummies(application_df)
dummies.head()
```

```python
# Split our preprocessed data into our features and target arrays
#  YOUR CODE GOES HERE

X = dummies.drop('IS_SUCCESSFUL',axis=1)
y=dummies['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
#  YOUR CODE GOES HERE

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
#  YOUR CODE GOES HERE
inputs = X_train_scaled.shape[1]
hidden_nodes1 =  200
hidden_nodes2 = 150
nn = tf.keras.models.Sequential()

# First hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes1, activation="relu", input_dim=inputs))


# Second hidden layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=hidden_nodes2, activation="relu"))


# Output layer
#  YOUR CODE GOES HERE
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))


# Check the structure of the model
nn.summary()
```

```python
# Compile the model
#  YOUR CODE GOES HERE
nn.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
# Train the model
#  YOUR CODE GOES HERE

model_fit = nn.fit(X_train_scaled,y_train,epochs=100)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
